# Elliptic Curves

An **elliptic curve** over a finite field $\mathbb{Z}_p$, where $p$ is prime, is the set of points

$$
(x, y) \in \mathbb{Z}_p^2
$$

satisfying the congruence

$$
y^2 \equiv x^3 + ax + b \pmod p
$$

where $a, b \in \mathbb{Z}_p$.

The curve must be **non-singular**, meaning that it has no cusps, self-intersections, or degenerate points. For the short Weierstrass equation above, this condition is

$$
4a^3 + 27b^2 \not\equiv 0 \pmod p.
$$

# Cyclic group from a elliptic curve

A set of object $G$ and a binary operation $\star: G\times G \rightarrow G$ form a **group** $(G, \star)$ when:
- $\star$ satisfy the **associativity** property, 
- there exists and **invariant/neuter element**,
- for each element in $G$ there exists an **inverse**.

A group is **cyclic** if there exists an element $g \in G$, called a **generator** or **primitive element**, such that every element of the group can be obtained by repeatedly applying the group operation to $g$.

$$\forall a \in G,\quad \exists k \in \mathbb{Z}_{|G|} \quad | \quad a = \underbrace{g \star g \star \cdots \star g}_{k \text{ times}} $$

The operation $\star$ is commonly interpreted as a "multiplication" such that the repeated application of the operation is interpreted as an **exponentiation**.
$$a = \underbrace{g \star g \star \cdots \star g}_{k \text{ times}} = g^k$$

However, it can also be interpreted as an "addition". In this case, the repeated application of the operation is interpreted as a **multiplication by a scalar**.
$$a = \underbrace{g \star g \star \cdots \star g}_{k \text{ times}} = k\cdot g$$

Here, we will use the second interpretation because is easier to deal with addition and multiplication by a scalar than multiplication and power.

Specifically, the inverse element with respect to addition can be difined by `__neg__` method, while the inverse element with respect to multiplication must be defined as the power by -1.

As a result, we will use $\star = +$ so that:
- the inverse element of $P = (x, y) \in G$ il $-P = (x, -y \bmod p)$
- the neuter element is $\infty = P + (-P)$

Hence, if $G$ is the set of point of an elliptic curve and $P = (x, y) \in G$, then the inverse element $\bar{P} = (x, -y \bmod p)$ as $-P$.
